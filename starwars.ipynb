```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36 Edg/88.0.705.50',
           'Referer':'https://www.starwars.com/databank'}
```

```python
url = 'https://www.starwars.com/_grill/filter/databank?filter=Characters&mod=3&offset={}'
```

```python
df = pd.DataFrame(columns=['Name', 'Pic-url', 'Description', 'Link', 'Appearances'])
for offset in range(0,840,40):
    print(offset , '-------------------------------------------------------------------------------------')
    res = requests.get(url.format(str(offset)), headers)
    j = json.loads(res.text)
    character_list = j['data']

    

    for character in character_list:
        name = character['title']
        pic_url = character['default_thumb']
        desc = character['desc']
        link = character['href']
        print(name)
        print(pic_url)
        print(desc)
        print(link)

        char = requests.get(link, headers)
        soup = BeautifulSoup(char.text, 'html.parser')
        if len(soup.select('div[class="category"]')) == 0 or soup.select('div[class="category"]')[0].text[0:4] != '\nApp':
            pass
        else:
            appear = soup.select('div[class="category"]')[0]
            tmp_appear_list = appear.text.replace('Appearances', '').split(',')
            appear_list = []
            for movie in tmp_appear_list[:-1]:
                appear_list.append(movie.strip())
            print(appear_list)

            char = pd.DataFrame([[name, pic_url, desc, link, appear_list]], columns=['Name', 'Pic-url', 'Description', 'Link', 'Appearances'])
            df = df.append(char, ignore_index=True)
```

```python
df.to_csv('starwars_char.csv')
```
